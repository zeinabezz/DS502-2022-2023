# <font color=blue>**Returning a new class object using class method**

````python
from datetime import date


class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def from_birth_year(cls, name, birthYear):
        return cls(name, date.today().year - birthYear)

p1=Person('ahmed', 30)
print(p1.name, p1.age)

p2=Person.from_birth_year('mohamed', 2000)
print(p2.name, p2.age)
````

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def from_full_name(cls, first_name, middle_name, last_name, age):
        return cls(first_name+' '+middle_name+' '+last_name, age)

p1=Person('ahmed', 30)
print(p1.name, p1.age)

p2=Person.from_full_name('mohamed', 'ahmed', 'ali',20)
print(p2.name, p2.age)
````

## <font color=blue>Difference between Class method, instance method, regular method?

class method:
- bounded method to the class 
- can only called from a class or class instance

instance method:
- bounded method to a class instance
- can only be called from a class instance

regular method:
- unbounded method
- can only be called from class 


## <font color=blue>When can I define class method, instance method, regular method?

class method:
- is used when there is a need to access the class attributes only [remember: it cannot access any instance attribute].
- can also be used creating a new class instance(object).

instance method: 
- is used when there is a need to access instance attributes [remember: it also can access class attributes and it can create a new class instance].

regular method: 
- can be used as a helping function that perform a particular task [remember, it can access class attributes].
- Thus it can be accessed by a class or an instance method

````python
class Circle:
  def __init__(self, radius=0.0):
    self.radius=radius
  
  @property
  def radius(self):
    return self._radius

  @radius.setter
  def radius(self, radius):
    if radius<0: print("Error")
    else: self._radius=radius


c=Circle(5)
c.radius=4
print(c.radius)

c2=Circle(10)
c2.radius=-3
print(c2.radius)
````

## <font color=red>Example

Given a point(x, y), Write Python Program to Find Whether it Lies in the First, Second, Third or Fourth Quadrant of x - y Plane


````python
class Quadrant:
  def __init__(self, cord_x, cord_y):
    self.cord_x=cord_x
    self.cord_y=cord_y

  @property
  def cord_x(self):
    return self._cord_x
  @cord_x.setter
  def cord_x (self, cord_x):
    self._cord_x=cord_x

  @property
  def cord_y(self):
    return self._cord_y
  @cord_y.setter
  def cord_y (self, cord_y):
    self._cord_y=cord_y


  def determine_quadrant(self):
    if self.cord_x >0 and self.cord_y>0 :  return "first quard"
    if self.cord_x <0 and self.cord_y>0 :  return "second quard"
    if self.cord_x <0 and self.cord_y<0 :  return "third quard"
    if self.cord_x >0 and self.cord_y<0:   return "fourth quad"


q1=Quadrant(2, -4)
print(q1.determine_quadrant())
````

# <font color=yellow> Inheritance
- Inheritance is a way to express is-a relationship between classes. 

- For example, a Manager is-a human, which is a living organism. postgraduate student is-a college student which is a student. 

- If the is-a relationship does not exist between a derived class and base class, you should not use inheritance.

- A class that is used as the basis for inheritance is called a superclass or base class or parent class. 

- A class that inherits from a base class is called a subclass or derived class or child class. 

- A derived class inherits variables and methods from its base class while adding additional variables and methods of its own. 

- Inheritance easily enables reusing of existing code.

- object class is the base of the inheritance hierarchy.

- All classes inheret implicitly from the object class

##<font color=green> Syntax of child class definintion:

       class SubClassName (SuperClassName):

<font color=orange>Thus,
     
       class x:
          statement(s)     

   <font color=orange> Is equivalent to 

      class x (object):
         statement(s)

# <font color=blue>Calling the parent class constructor using </font>super().\_\_init\_\_(...)

# <font color=reyellow> Multiple Inheritance

#Python also supports a form of multiple inheritances. 

- A derived class definition with multiple base classes looks like this:

           class DerivedClassName(Base_1, Base_2, Base_3):
                      <statement-1>
                          . . .
                     <statement-N>

- Derived class is inherited from multiple base classes, Base_1, Base_2, Base_3. 

- For most purposes, in the simplest cases, you can think of the search for attributes inherited from a parent class as depth-first, left-to-right, not searching twice in the same class where there is an overlap in the hierarchy. 

- Thus, if an attribute is not found in Derived Class, it is searched for in Base_1, then (recursively) in the base classes of Base_1, and if it was not found there, it would be searched for in Base_2, and so on. 

- Even though multiple inheritances are available in Python programming language, it is not highly encouraged to use it as it is hard and error prone.

- You can call the base class method directly using Base Class name itself without using the super() function.  The syntax is,

            BaseClassName.methodname(self, arguments)


````python
class LivingOrganism:
  def __init__(self, can_eat, organ_type):
    self.organ_type=organ_type
    self.can_eat=can_eat

  @property
  def can_eat(self):
    return self._can_eat

  @can_eat.setter
  def can_eat(self, can_eat):
    self._can_eat= can_eat

  @property
  def organ_type(self):
    return self._organ_type

  @organ_type.setter
  def organ_type(self, organ_type):
    self._organ_type= organ_type


class Human (LivingOrganism):
  def __init__(self, name):
    super().__init__(True, 'human')
    self.name=name

    @property
    def name(self):
       return self._name

    @name.setter
    def name(self, name):
        self._name= name


class Employee(Human):
  def __init__(self, name, emp_dept):
      super().__init__(name)
      self.emp_dept=emp_dept

  @property
  def emp_dept(self):
       return self._emp_dept

  @emp_dept.setter
  def emp_dept(self, emp_dept):
        self._emp_dept= emp_dept

  def display_info(self):
    print("From Employee Class: ", self.name, self.emp_dept)

class Student(Human):
  def __init__(self, name, st_dept):
    Human.__init__(self, name)
    self.st_dept=st_dept

  @property
  def st_dept(self):
       return self._st_dept

  @st_dept.setter
  def st_dept(self, st_dept):
        self._st_dept= st_dept

  def display_info(self):
    print("From Student Class: ", self.name, self.st_dept)

class PostGraduateStudent(Student, Employee):
        def __init__(self, name, st_dept, emp_dept):
          Student.__init__(self, name, st_dept)
          Employee.__init__(self, name, emp_dept)

        
se1=PostGraduateStudent('ahmed', 'DS', 'maintenance')
se1.display_info()
````

# <font color=blue> issubclass()
- To check the class inheritance use:  **issubclass() **
          
          issubclass(SubClassName, SuperClassName)

- This function returns Boolean True if SubClassName is a subclass of superclass SuperClassName.

- The SubClassName class is considered a subclass of itself. 

- SuperClassName may be a tuple of classes, in which case every entry in SuperClassName will be checked. 

- In any other case, a TypeError exception is raised.


````python
class LivingOrganism:
  def __init__(self, can_eat, organ_type):
    self.organ_type=organ_type
    self.can_eat=can_eat

  @property
  def can_eat(self):
    return self._can_eat

  @can_eat.setter
  def can_eat(self, can_eat):
    self._can_eat= can_eat

  @property
  def organ_type(self):
    return self._organ_type

  @organ_type.setter
  def organ_type(self, organ_type):
    self._organ_type= organ_type


class Human (LivingOrganism):
  def __init__(self, name):
    super().__init__(True, 'human')
    self.name=name

    @property
    def name(self):
       return self._name

    @name.setter
    def name(self, name):
        self._name= name


class Employee(Human):
  def __init__(self, name, emp_dept):
      super().__init__(name)
      self.emp_dept=emp_dept

  @property
  def emp_dept(self):
       return self._emp_dept

  @emp_dept.setter
  def emp_dept(self, emp_dept):
        self._emp_dept= emp_dept

  def display_info(self):
    print("From Employee Class: ", self.name, self.emp_dept)

class Student(Human):
  def __init__(self, name, st_dept):
    Human.__init__(self, name)
    self.st_dept=st_dept

  @property
  def st_dept(self):
       return self._st_dept

  @st_dept.setter
  def st_dept(self, st_dept):
        self._st_dept= st_dept

  def display_info(self):
    print("From Student Class: ", self.name, self.st_dept)

class PostGraduateStudent(Student, Employee):
        def __init__(self, name, st_dept, emp_dept):
          Student.__init__(self, name, st_dept)
          Employee.__init__(self, name, emp_dept)

        
se1=PostGraduateStudent('ahmed', 'DS', 'maintenance')
print(issubclass(PostGraduateStudent, Employee))
print(issubclass(PostGraduateStudent, (Employee, Student)))
````

# <font color=yellow> Polymorphism
- Polymorphism means that you can have multiple classes where each class implements the same variables or methods in different ways.

- A real-world example of polymorphism is suppose when if you are in classroom that time you behave like a student, when you are in market at that time you behave like a customer, when you are at your home at that time you behave like a son or daughter, such that same person is presented as having different behaviors.


    overload methods: methods with the same name but with different parameter list
    override methods: methods with the same name and the same parameter list  [in case of inheritance]
    overload methods are developed with *args or **kwargs.

````python
import math
class Shape:
    def area(self):
        pass
    def perimeter(self):
        pass


###
class Rectangle(Shape):
    def __init__(self, width, height):
            self.width = width
            self.height = height

    def area(self):
         print(f"Area of Rectangle is {self.width * self.height}")

    def perimeter(self):
        print(f"Perimeter of Rectangle is {2 * (self.width + self.height)}")



###
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        print(f"Area of Circle is {math.pi * self.radius ** 2}")

    def perimeter(self):
        print(f"Perimeter of Circle is {2 * math.pi * self.radius}")


###
def shape_type(shape_obj):
    shape_obj.area()
    shape_obj.perimeter()

def main():
    rectangle_obj = Rectangle(10, 20)
    circle_obj = Circle(10)
    for each_obj in [rectangle_obj, circle_obj]:
        shape_type(each_obj)
main()
````

# <font color=red> Example on overload methods

````python
class Person:
       
    def __init__(self, myname, myage, *myaddress):
        self._name=myname
        self._age=myage
        self._address=""
        for address in myaddress:
           address+=", "
           self._address= self._address + address


    @property
    def name(self):
        return self._name
    @property
    def age(self):
        return self._age
    @property
    def address(self):
        return self._address
    

    @name.setter    
    def name(self, name):
        self._name=name
        
    @age.setter    
    def age(self, age):
        self._age=age
    @address.setter    
    def address(self, address):
        self._address=address
   



p1=Person("ahmed", 23, "Giza")
print(p1.address)
p2=Person("mohammed", 24, "Giza", "Ahmed Zewail st.")
print(p2.address)
p3=Person("mahmoud", 25, "Giza", "Ahmed Zewail st.", "Cairo Uni.")
print(p3.address)
````
